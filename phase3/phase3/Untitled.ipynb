```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Lire le fichier CSV
df = pd.read_csv("resultats3.csv")

# Axe X = nombre de threads T
x = np.arange(len(df["T"]))

# Largeur des barres
largeur = 0.35

# Créer la figure
plt.figure(figsize=(10, 6))

# Barres pour temps_calcul
plt.bar(x - largeur/2, df["temps_calcul"], width=largeur, label="Temps calcul")

# Barres pour temps_execution
plt.bar(x + largeur/2, df["temps_execution"], width=largeur, label="Temps exécution")

# Ajouter les labels et titre
plt.xlabel("Nombre de threads (T)")
plt.ylabel("Temps (secondes)")
plt.title("Comparaison des temps de calcul et d'exécution")
plt.xticks(x, df["T"])  # afficher les valeurs de T sur l'axe X
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Afficher le graphe
plt.tight_layout()
plt.show()
```
